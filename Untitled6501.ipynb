```python
import os
import numpy as np
from PIL import Image
from keras.applications.vgg19 import VGG19, preprocess_input
from keras.models import Model
from keras.layers import Dense, GlobalAveragePooling2D
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split

# Especificar o caminho para o diretrio que contm as imagens
diretorio_imagens = '/content/GroceryStoreDataset/dataset/train/Fruit/Banana'

# Listar todos os arquivos no diretrio de imagens
imagens = os.listdir(diretorio_imagens)

# Pr-processamento e carregamento das imagens
X = []
y = []

for imagem_nome in imagens:
    # Construir o caminho completo para a imagem
    caminho_imagem = os.path.join(diretorio_imagens, imagem_nome)

    img = Image.open(caminho_imagem)
    img = img.resize((224, 224))  # Redimensionar para o tamanho esperado pela VGG19
    img = np.array(img)
    img = preprocess_input(img)  # Pr-processamento especfico da VGG19

    X.append(img)
    y.append(imagem_nome)  # Nome da imagem como rtulo

X = np.array(X)
y = np.array(y)

# Codificar os rtulos em nmeros
label_encoder = LabelEncoder()
y_encoded = label_encoder.fit_transform(y)

# Dividir o conjunto de dados em treinamento e teste
X_train, X_test, y_train, y_test = train_test_split(X, y_encoded, test_size=0.2, random_state=42)

# Carregar o modelo VGG19 pr-treinado com pesos do ImageNet
base_model = VGG19(weights='imagenet', include_top=False)

# Adicionar camadas personalizadas para a tarefa de classificao
x = base_model.output
x = GlobalAveragePooling2D()(x)
x = Dense(1024, activation='relu')(x)
predictions = Dense(len(label_encoder.classes_), activation='softmax')(x)

# Criar o modelo final
model = Model(inputs=base_model.input, outputs=predictions)

# Descongelar algumas camadas convolucionais para ajuste fino
for layer in base_model.layers[:15]:
    layer.trainable = True

# Compilar o modelo
model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])

# Treinar o modelo com ajuste fino
model.fit(X_train, y_train, epochs=10, batch_size=32, validation_split=0.2)

# Avaliar o modelo no conjunto de teste
accuracy = model.evaluate(X_test, y_test)[1]
print(f'Acurcia no conjunto de teste: {accuracy * 100:.2f}%')
```

```python
import os
import cv2
import numpy as np
from keras.applications.vgg19 import VGG19, preprocess_input, decode_predictions
from keras.preprocessing.image import load_img, img_to_array
from sklearn.metrics import accuracy_score

# Caminho para o diretrio de imagens de treinamento
diretorio_imagens_treinamento = '/content/GroceryStoreDataset/dataset/train/Fruit/Banana'

# Caminho para o diretrio de imagens de teste
diretorio_imagens_teste = '/content/GroceryStoreDataset/dataset/test/Fruit/Banana'

# Carregar o modelo VGG19 pr-treinado
model = VGG19(weights='imagenet')

# Funo para realizar a classificao de imagem
def classify_image(image_path, model):
    # Carregar a imagem
    image = load_img(image_path, target_size=(224, 224))
    image = img_to_array(image)
    image = np.expand_dims(image, axis=0)
    image = preprocess_input(image)

    # Realizar a classificao de imagem
    predictions = model.predict(image)

    return predictions

# Lista de imagens no diretrio de teste
imagens_teste = os.listdir(diretorio_imagens_teste)

# Suponhamos que voc queira avaliar a preciso do modelo para a classe "chocolate"
classe_alvo = "chocolate"
precisoes = []

# Loop atravs de todas as imagens de teste
for imagem_teste in imagens_teste:
    caminho_imagem_teste = os.path.join(diretorio_imagens_teste, imagem_teste)
    predictions_teste = classify_image(caminho_imagem_teste, model)
    decoded_predictions_teste = decode_predictions(predictions_teste, top=5)[0]

    # Obter a classe com a maior confiana
    class_id_teste, class_label_teste, class_score_teste = decoded_predictions_teste[0]

    # Verificar se a classe predita corresponde  classe alvo
    if class_label_teste.lower() == classe_alvo:
        precisao = 1.0
    else:
        precisao = 0.0

    precisoes.append(precisao)

# Calcular a preciso mdia
precisao_media = sum(precisoes) / len(precisoes)

# Exibir o resultado da preciso mdia
print(f'Preciso Mdia para a Classe "{classe_alvo}": {precisao_media:.2f}')
```

```python
from IPython.display import display, Javascript
from google.colab.output import eval_js
from base64 import b64decode
from pyzbar.pyzbar import decode
import cv2
import numpy as np

def capture_barcode():
    js = Javascript('''
    async function captureBarcode() {
      const div = document.createElement('div');
      const capture = document.createElement('button');
      capture.textContent = 'Capture';
      div.appendChild(capture);

      const video = document.createElement('video');
      video.style.display = 'block';
      const stream = await navigator.mediaDevices.getUserMedia({video: true});

      document.body.appendChild(div);
      div.appendChild(video);
      video.srcObject = stream;
      await video.play();

      // Resize the output to fit the video element.
      google.colab.output.setIframeHeight(document.documentElement.scrollHeight, true);

      // Wait for Capture to be clicked.
      await new Promise((resolve) => capture.onclick = resolve);

      const canvas = document.createElement('canvas');
      canvas.width = video.videoWidth;
      canvas.height = video.videoHeight;
      canvas.getContext('2d').drawImage(video, 0, 0);
      stream.getVideoTracks()[0].stop();
      div.remove();
      return canvas.toDataURL('image/jpeg');
    }
    ''')

    display(js)
    data = eval_js('captureBarcode()')
    binary = b64decode(data.split(',')[1])

    # Salvar a imagem temporariamente
    with open('temp.jpg', 'wb') as f:
        f.write(binary)

    # Ler o cdigo de barras na imagem
    image = cv2.imread('temp.jpg', 0)
    decoded_objects = decode(image)
    if decoded_objects:
        for obj in decoded_objects:
            barcode_data = obj.data.decode('utf-8')
            print('Cdigo de barras lido:', barcode_data)
    else:
        print('Nenhum cdigo de barras foi detectado.')

    # Remover a imagem temporria
    import os
    os.remove('temp.jpg')

capture_barcode()
```
